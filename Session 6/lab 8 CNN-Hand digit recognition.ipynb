# Data Preparation
## Reshaping the Image Data

```python
#CNN require images to be in the shape (width, height, channels)
#Reshape the X_train dataset
#Why channels is 1 in this case, answer using another # line
X_train = X_train.reshape((60000, 28, 28, 1))
```

The picture is only in black and white therefore the channel can only be one

```python
#Check the shape now, what has changed
X_train.shape
```

```python
#Now do the same for X_test
X_test = X_test.reshape((10000, 28, 28, 1))
```

```python
#Check the shape now
X_test.shape
```

# Normalizing the Image Data

```python
#Do you still remember what is normalisation?
#Write it here

#Now normalise the X_train
X_train = X_train.astype('float32') / 255
```

Normalisation is when you convert the values to your ratio between - 1  1 This is to scale each of the features so that none of the features is dominant over the other. This improves the performance and efficiency of the model and does not dominate one feature does the creation of the model of the model. 

```python
#Do the same for X_test
X_test = X_test.astype('float32') / 255
```

# One-Hot Encoding: Converting the Labels From Integers to Categorical Data

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_train.shape
```

```python
y_train[0]
```

```python
y_test = to_categorical(y_test)
y_test.shape
#Explain what has changed and why using a new # line
```

# Creating the Neural Network

```python
#Import sequential from keras models
from tensorflow.keras.models import Sequential
cnn = Sequential()
```

# Adding Layers to the Network

```python
from tensorflow.keras.layers import Conv2D, Dense, Flatten, MaxPooling2D
```

# Adding a Convolution Layer

```python
cnn.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
#Define activation function?
```

# Dimensionality of the First Convolution Layer’s Output

# Adding a Pooling Layer

```python
cnn.add(MaxPooling2D(pool_size=(2, 2)))
#What is max pooling techanique and why it is used?
```

# Adding Another Convolutional Layer and Pooling Layer

```python
cnn.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
```

# Flattening the Results

```python
cnn.add(Flatten())
```

# Adding a Dense Layer to Reduce the Number of Features

```python
cnn.add(Dense(units=128, activation='relu'))
```

# Adding Another Dense Layer to Produce the Final Output

```python
cnn.add(Dense(units=10, activation='softmax'))
```

# Printing the Model’s Summary

```python
cnn.summary()
```

# Visualizing a Model’s Structure

```python
from tensorflow.keras.utils import plot_model
from IPython.display import Image
plot_model(cnn, to_file='convnet.png', show_shapes=True,
            show_layer_names=True)
Image(filename='convnet.png')
```

# Compiling the Model

```python
cnn.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
```

# Training and Evaluating the Model

```python
cnn.fit(X_train, y_train, epochs=5, batch_size=64,
              validation_split=0.1)
```

# Evaluating the Model

```python
loss, accuracy = cnn.evaluate(X_test, y_test)
```

# Making Predictions

```python
predictions = cnn.predict(X_test)
```

```python
y_test[0]
```

```python
for index, probability in enumerate(predictions[0]):
          print(f'{index}: {probability:.10%}')
```

# Locating the Incorrect Predictions

```python
images = X_test.reshape((10000, 28, 28))
incorrect_predictions = []

for i, (p, e) in enumerate(zip(predictions, y_test)):
    predicted, expected = np.argmax(p), np.argmax(e)

    if predicted != expected:
        incorrect_predictions.append(
            (i, images[i], predicted, expected))
```

```python
len(incorrect_predictions)
```

# Visualizing Incorrect Predictions

```python
figure, axes = plt.subplots(nrows=4, ncols=6, figsize=(16, 12))

for axes, item in zip(axes.ravel(), incorrect_predictions):
    index, image, predicted, expected = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([]) # remove x-axis tick marks
    axes.set_yticks([]) # remove y-axis tick marks
    axes.set_title(
        f'index: {index}\np: {predicted}; e: {expected}')
plt.tight_layout()
```

# Displaying the Probabilities for Several Incorrect Predictions

```python
def display_probabilities(prediction):
    for index, probability in enumerate(prediction):
        print(f'{index}: {probability:.10%}')
```
